## Kota Bandung (district) analysis

```python
bandung_flood_df_data = []
for year in range(2020, 2025 + 1):
    for month in range(1, 12 + 1):
        filtered_df = bandung_df[bandung_df["Tanggal Kejadian"].map(lambda x: x.year == year and x.month == month)]
        flood = len(filtered_df) > 0
        bandung_flood_df_data.append({
            "year": year,
            "month": month,
            "flood": flood
        })

bandung_flood_df = pd.DataFrame(bandung_flood_df_data)
bandung_flood_df
```

```python
modified_bandung_flood_df = bandung_flood_df.copy()
```

```python
modified_bandung_flood_df["year_month_id"] = modified_bandung_flood_df["year"] * 12 + modified_bandung_flood_df["month"]
modified_bandung_flood_df
```

```python
import seaborn as sns
```

```python
sns.lineplot(modified_bandung_flood_df, x="year_month_id", y="flood")
```

Jawa Barat and JAWA BARAT -> This is a same thing!

```python
jabar_df = df[df["Provinsi"].map(lambda x: x.lower() == "jawa barat")].copy()
jabar_df = jabar_df.drop(columns=["Provinsi", "Kronologi & Dokumentasi"])
jabar_df
```

```python
jabar_df.drop_duplicates(["ID Kabupaten", "Kabupaten"])
```

```python
jabar_df["Kejadian"].value_counts()
```

```python
jabar_df["Tanggal Kejadian"].max()
```

```python
jabar_df["Tanggal Kejadian"].min()
```

We make a new table by counting for each date. We skipped December 2025 because that month has incomplete information.

```python
from collections import Counter

jabar_event_df_data = []
for current_x in pd.date_range(start='2020-01-01', end='2025-11-30'):
    filtered_df = jabar_df[jabar_df["Tanggal Kejadian"] == current_x]
    result = {"date": current_x, "year": current_x.year, "month": current_x.month}
    filtered_df = filtered_df.drop_duplicates(subset=["ID Kabupaten", "Kejadian"])
    for event in filtered_df["Kejadian"].unique():
        result[event] = len(filtered_df[filtered_df["Kejadian"] == event])
    jabar_event_df_data.append(result)

jabar_event_df = pd.DataFrame(jabar_event_df_data)
jabar_event_df = jabar_event_df.fillna(0)
jabar_event_df
```

```python
grouped_df = jabar_event_df.drop(columns=["date"]).groupby(["year", "month"]).sum().reset_index()
grouped_df["date"] = pd.to_datetime(grouped_df.apply(lambda x: f"{int(x['year'])}-{int(x['month']):0>2}-01", axis=1))
grouped_df
```

```python
import matplotlib.pyplot as plt
```

```python
# sns.lineplot(grouped_df, x="date", y="CUACA EKSTREM", label="CUACA EKSTREM")
# sns.lineplot(grouped_df, x="date", y="TANAH LONGSOR", label="TANAH LONGSOR")
sns.lineplot(grouped_df, x="date", y="BANJIR")
plt.title("Flood Trend in Jawa Barat")
```

We make it simple because ... if there's a lot of lines, it can make the reader more confused.

Insight: The changing-year moment is always give a significant peak, but year 2023 and 2024 seems not extreme, unlike the previous years. However, that changes in 2025.

## Indonesia (country) analysis

```python
event_df_data = []
for current_x in pd.date_range(start='2020-01-01', end='2025-11-30'):
    filtered_df = df[df["Tanggal Kejadian"] == current_x]
    result = {"date": current_x, "year": current_x.year, "month": current_x.month}
    filtered_df = filtered_df.drop_duplicates(subset=["ID Kabupaten", "Kejadian"])
    for event in filtered_df["Kejadian"].unique():
        result[event] = len(filtered_df[filtered_df["Kejadian"] == event])
    event_df_data.append(result)

event_df = pd.DataFrame(event_df_data)
event_df = event_df.fillna(0)
event_df
```

```python
grouped_df = event_df.drop(columns=["date"]).groupby(["year", "month"]).sum().reset_index()
grouped_df["date"] = pd.to_datetime(grouped_df.apply(lambda x: f"{int(x['year'])}-{int(x['month']):0>2}-01", axis=1))
grouped_df
```

```python
# sns.lineplot(grouped_df, x="date", y="CUACA EKSTREM", label="CUACA EKSTREM")
# sns.lineplot(grouped_df, x="date", y="TANAH LONGSOR", label="TANAH LONGSOR")
sns.lineplot(grouped_df, x="date", y="BANJIR")
plt.title("Flood Trend in Indonesia")
```

Although the order is somehow awkward, it's positively correlated with the number of districts in a province.

```python
from tqdm import tqdm
```

```python
province_event_df_data = []
province_list = df["Provinsi"].unique()

for current_x in tqdm(pd.date_range(start='2020-01-01', end='2025-11-30')):
    for province in province_list:
        filtered_df = df[(df["Tanggal Kejadian"] == current_x) & (df["Provinsi"] == province)]
        result = {"date": current_x, "year": current_x.year, "month": current_x.month, "province": province}
        filtered_df = filtered_df.drop_duplicates(subset=["ID Kabupaten", "Kejadian"])
        for event in filtered_df["Kejadian"].unique():
            result[event] = len(filtered_df[filtered_df["Kejadian"] == event])
        province_event_df_data.append(result)

province_event_df = pd.DataFrame(province_event_df_data)
province_event_df = province_event_df.fillna(0)
province_event_df
```

```python
province_event_df.columns
```

```python
province_event_df = province_event_df.astype({'BANJIR': "int", 'TANAH LONGSOR': "int",
       'CUACA EKSTREM': "int", 'KEBAKARAN HUTAN DAN LAHAN': "int", 'GEMPABUMI': "int",
       'GELOMBANG PASANG DAN ABRASI': "int", 'KEKERINGAN': "int", 'ERUPSI GUNUNG API': "int",
       'TSUNAMI': "int"})
province_event_df
```

```python
province_event_df.to_csv("province_events.csv", index=False)
```
