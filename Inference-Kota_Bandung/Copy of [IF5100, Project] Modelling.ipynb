```python
from sklearn.linear_model import LogisticRegression
```

```python
df.head()
```

```python
kecamatan_list = df["Kecamatan"].unique().tolist()
kecamatan_list.sort()
```

```python
import numpy as np
from tqdm import tqdm

data_matrix = []

for current_x in tqdm(pd.date_range(start=df["Tanggal Kejadian"].min(), end=df["Tanggal Kejadian"].max())):
    filtered_df = df[df["Tanggal Kejadian"] == current_x]
    if filtered_df.empty:
        continue

    row = []
    for kecamatan in kecamatan_list:
        if filtered_df[filtered_df["Kecamatan"] == kecamatan].empty:
            row.append(0)
        else:
            row.append(1)

    data_matrix.append(row)

data_matrix = np.array(data_matrix)
data_matrix.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_matrix_train, data_matrix_test = train_test_split(data_matrix, test_size=0.2, random_state=120)
data_matrix_train.shape, data_matrix_test.shape
```

```python
clf_list = []
for i in tqdm(range(len(kecamatan_list))):
    X = data_matrix_train[:,[j for j in range(len(kecamatan_list)) if j != i]]
    y = data_matrix_train[:,i]
    if len(np.unique(y)) == 1:
        clf_list.append({
            "type": "constant",
            "model": y[0]
        })
    else:
        clf = LogisticRegression(random_state=120)
        clf.fit(X, y)
        clf_list.append({
            "type": "logreg",
            "model": clf
        })

len(clf_list)
```

```python
count = 0
for clf_data in clf_list:
    if clf_data["type"] == "logreg":
        count += 1

count
```

```python
all_y = []
all_y_pred = []
for i, clf_data in tqdm(enumerate(clf_list), total=len(clf_list)):
    X = data_matrix_test[:,[j for j in range(len(kecamatan_list)) if j != i]]
    y = data_matrix_test[:,i]
    all_y.extend(y)

    if clf_data["type"] == "constant":
        y_pred = [clf_data["model"] for _ in X]
    else:
        clf = clf_data["model"]
        y_pred = clf.predict(X)
    all_y_pred.extend(y_pred)

len(all_y), len(all_y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(all_y, all_y_pred))
```
